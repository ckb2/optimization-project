```r
library(tidyverse)
library(rvest)
library(stringr)

url <- "https://www.myplate.gov/recipes/supplemental-nutrition-assistance-program-snap/2-step-chicken"

suppressMessages(test_recipe <- read_html(url))
```

```r
# Ingredients
test_recipe %>%
  html_nodes("li.field__item") %>%
  html_text() %>%
  str_squish() %>%
  tibble()
```

```r
# Servings
test_recipe %>%
  html_node(".mp-recipe-full__detail--data , .yyyyy") %>%
  html_text() %>%
  str_squish()
```

```r
# Nutritional info
test_recipe %>%
  html_node(".mp-recipe-full__nutrition-form") %>%
  html_table()
```
