```r
library(tidyverse)
library(rvest)
library(stringr)
```

```r
get_links <- function() {
    base_url <- "https://www.myplate.gov/myplate-kitchen/recipes?items_per_page=100&sort_bef_combine=title_ASC&sort_by=title&sort_order=ASC&page="
    
    link_list <- list()    
    for (i in 0:10) {
        link_list <- c(link_list,
                read_html(paste0(base_url, i)) %>%
                    html_nodes("a") %>%
                    html_attr("href") %>%
                    str_subset("\\/recipes\\/")
                )
    }

    return(unique(link_list))
}

get_recipe <- function(url) {
    suppressMessages(recipe <- read_html(url))

    title <- recipe %>%
    html_node(".field--label-hidden") %>%
    html_text() %>%
    str_squish()

    ingredients <- recipe %>%
        html_nodes("li.field__item") %>%
        html_text() %>%
        str_squish() %>%
        tibble()

    servings <- recipe %>%
        html_node(".mp-recipe-full__detail--data , .yyyyy") %>%
        html_text() %>%
        str_squish() %>%
        str_extract("[0-9]+")

    directions <- recipe %>%
        html_node("ol") %>%
        html_text() %>%
        str_squish()

    nutritional_info <- recipe %>%
        html_node(".mp-recipe-full__nutrition-form") %>%
        html_table()

    food_groups <- recipe %>%
        html_node(".mp-food-group") %>%
        html_table()

    return(tibble_row('Name' = list(title), 'Link' = url, 'Ingredients' = list(ingredients), 'Directions' = list(directions), 'Servings' = servings, 'FoodGroups' = list(food_groups), 'NutritionalInfo' = list(nutritional_info)))
}
```

```r
links <- get_links()
links
```

```r
my_plate_recipes <- tibble()
base_url <- "https://www.myplate.gov"

for (link in links) {
    url <- paste0(base_url, link)
    try(my_plate_recipes <- my_plate_recipes %>% bind_rows(get_recipe(url)))
    Sys.sleep(1)
}
```

```r
my_plate_recipes %>%
    saveRDS("my_plate_recipes.rds")
    
```
